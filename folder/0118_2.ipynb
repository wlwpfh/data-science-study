### K-fold  


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# 5개의 폴드 세트로 분리하는 KFold 객체와 폴드 세트별 정확도를 담을 리스트 객체 생성.
kfold = KFold(n_splits=5)
# 몇 등분을 할 것인지 

cv_accuracy = []
print('붓꽃 데이터 세트 크기:',features.shape[0])
```

```python
kfold.split(iris.data)
```

```python
iris.data
```

```python
list(kfold.split(iris.data))

# 5등분중에 4개를 학습하고 남은 하나 
```

```python
n_iter=0

for idx_train, index_test in kfold.split(iris.data):
    print(idx_train, index_test)
```

```python
n_iter=0
cv_acc=0

for idx_train, idx_test in kfold.split(iris.data):
    X_train, X_test= iris.data[idx_train], iris.data[idx_test]
    y_train, y_test= iris.target[idx_train], iris.data[idx_test]
    
    dt_clf.fit(X_train, y_train)
    pred=dt_clf.predict(X_test)
    n_iter+=1
    size_train=X_train.shape[0]
    size_test=X_test.shape[1]
    
    accuracy= accuracy_score(y_test, pred)
    cv_acc+=accuracy
    
```

### cross_val_score

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score , cross_validate
from sklearn.datasets import load_iris
import numpy as np

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)

data = iris_data.data
label = iris_data.target
```

```python
cross_val_score(dt_clf, data, label, cv=5, scoring='accuracy')

# 어던 것을 돌릴 것인지, feature와 label이 뭔지, cv옵션으로 몇 번 돌릴 것인지
```
