#### 1~81페이지까지 자동화하는 방법 
-> 자동으로 스크롤해주는 기능이 있다.
1) 1페이지부터 시작  
 ( xpath를 이용해서 1-10을 수행하고 10의 배수일 때 >의 xpath찾아서 넘기기  
2) driver 실행 -> html을 받음  
3) html parsing 후 find하기 (해당 시, 구역 찾기)  
4) 2페이지로 이동하기 ~ 무한 반복  


```python
# xpaht를 찾는 방법: 사용자도구를 통해서 copy xpath를 통해서 값 받기 
xpath = """//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/span/a[10]"""
driver.find_element_by_xpath(xpath).click()
```

```python
//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/span/a[1]
//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/span/a[3]
//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/span/a[10]

# xpath의 생김새
# 11이어도 1로 나온다. -> 1~10이 순환적으로 된다. 
# 화살표키도 버튼키인데 xpath를 따로 봐야 한다. (이건 고정값)
```

```python
list_mcd_addr = []

max_page = 81     #현재 기준 McDonalds 매장 주소화면 마지막 page 기준

for p in range(1,max_page+1):
    if p % 10 == 0:
        i = 10
    else:
        i = p % 10
    xpath = f"""//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/span/a[{i}]"""
    driver.find_element_by_xpath(xpath).click()
    time.sleep(1)
    
    # 주소정보 Collect : 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    addr_raw = soup.find_all('dd', 'road')
    
    for n in addr_raw:
        if '서울' in n.text[0:3]: #서울시의 
            list_mcd_addr.append(n.text.split(' ')[1]) # 구만 골라내기

    # Next page 검사 : 
    if p % 10 == 0: 
        xpath = """//*[@id="container"]/div[1]/div[2]/div/div/div[2]/div/a[3]"""   # Next page
        driver.find_element_by_xpath(xpath).click()
        time.sleep(1)
        # 오른쪽 화살표를 통해서 다음 배수의 페이지들로 이동
        
```
