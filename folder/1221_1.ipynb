```python
# Data 로딩 (xlsx 파일)
population = pd.read_excel('population_raw_data.xlsx', header=1)

# header 1을 해서 행정구역 및 나이대가 헤더로 간다. 

population.head(4)

#엑셀에서 하나의 index가 3줄을 포함하기 때문에 NaN값으로 나온다. 
# -> 전처리 하기 
```

```python
population.fillna(method='pad', inplace=True)    
# ffill: 결측치NaN 앞의 값으로 채우기,
# inplace: 업데이트까지 수행 옵션
# 상황에 따라 값을 넣고 싶을 때: pad (ffill, backfill, 등등이 있음)
# -> 위에와 동일한 값들이 생략된거니까 하나로 퉁친걸 ㄱ알고 pad임. 다음 값이 나오기 전꺼ㅏ지
# 위의 경우처럼 규칙성이 있는 경우 

# 결측값을 앞 방향 혹은 뒷 방향으로 채우기 (fill gaps forward or backward)

#      : fillna(method='ffill' or 'pad'), fillna(method='bfill' or 'backfill')



#출처: https://rfriend.tistory.com/262 [R, Python 분석과 프로그래밍의 친구 (by R Friend)]
```

```python
# 2. 컬럼명 단순화
population.rename(columns = {'행정구역(동읍면)별(1)':'광역시도', '행정구역(동읍면)별(2)':'시도','계':'인구수','항목':'구분'}, inplace=True)

population.columns
```

```python
# 인구 소멸 지수 웃기네...
# 만 20-39세 여성 인구 수 / 만 65세 이상 고령 인구 수 
```

```python
# 데이터 정비 :

# 소계를 제외하고 보고싶기 때문에 -> 소계를 제외하면 구 단위로 볼 수 있음.
population[population['시도'] != '소계']
```

```python
# 구분 칼럼의 있는 값들을 컨트롤하기 쉽게 바꾸기 
# '총인구수(명) -> '합계'
# '남자인구수(명) -> '남자'
# '여자인구수(명) -> '여자'
population.loc[population['구분'] == '총인구수 (명)', '구분'] = '합계'
population.loc[population['구분'] == '남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분'] = '여자'

# column의 이름 바꾸기 
```

```python
population.head(10)
population[population['시도'] != '소계'].head(10)
```

```python
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                        population['30 - 34세'] + population['35 - 39세']
    
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세'] + population['85 - 89세'] + \
                        population['90 - 94세'] + population['95 - 99세'] + population['100+']

# 일단 연령을 기준으로 필요한 정보들 합산 

# 새로운 컬럼명이기 때문에 새로 생김 
```

```python
population.head(10)
population[population['시도'] != '소계'].head(10)
```

```python
pop = pd.pivot_table(population, index=['광역시도','시도'], values=['인구수','20-39세','65세이상'])
pop

# column이 없기 때문에 합해서 평균값을 낸다. (남자 여자 값을 독립적으로 판단해서)
```

```python
pop = pd.pivot_table(population, index=['광역시도','시도', '구분'], values=['인구수','20-39세','65세이상'])
pop

# 아니면 index에 구분을 넣어도 된다. 
```

```python
pop = pd.pivot_table(population, index=['광역시도','시도'], columns='구분', values=['인구수','20-39세','65세이상'])
pop

# index를 key값으로 해서 묶는다. values를 통해서 정의함.
# columns는 구분값을 넣으면 column을 기준으로 또 분류한다. (남자, 여자, 합계가 나옴)
```

```python
pop['소멸비율'] = pop['20-39세','여자'] / pop['65세이상','합계']
```

```python
pop.head()
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 0.5     # True: 소멸위기지역,  False: 소멸위기지역 아님
pop.head()

# 위험하다 아니다를 판단할 수 있음. 
```
