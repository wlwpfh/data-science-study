# Countplot
Countplot은 데이터의 개수를 나타내는 그래프입니다. 자체적으로 여러 옵션을 지정해줄 수 있음. x축이나 y축에 놓을 칼럼을 지정할 수 있다.

```python
sns.countplot(data=data, x='성별코드')

# 차이를 보려고 하는 것이 아님. 직관적인 인식을 받기 위한 것 

# 성별코드를 기준으로 보는 것 :1아니면 2로 된 데이터이다. 1과 2를 x축에 넣은 것 
```

```python
sns.countplot(data=data, y='성별코드')

# y축으로 본 경우 
```

# Displot
데이터의 분포, 데이터의 편파성

연속적으로 표현된다. 

-> 근데 곧 없어지니 histplot으로 사용하면 된다. 

```python
sns.histplot(data['허리둘레'], kde=True)

#함수가 업데이트가 되어서 그런 것
```

```python
one = data[data['흡연상태'] == 1]
two = data[data['흡연상태'] == 2]
three = data[data['흡연상태'] == 3]
```

```python
plt.figure(figsize=(9, 3))
sns.distplot(one['허리둘레'], hist=False, label='흡연 안 함')
sns.distplot(two['허리둘레'], hist=False, label='흡연 했지만 그만 둠')
sns.distplot(three['허리둘레'], hist=False, label='흡연 중')
```
