```python

## 파일 읽어오기 :
data = pd.read_csv("weight_log.csv",encoding='utf-8')  #추가 주요옵션: encoding=utf-8/cp949/euc-kr, header=[0,1], index_col = "회차"
data
```

파이프라이닝 - 함수를 이어서 쓸 수 있음

```python
data["지점"] = data["지점"].replace("강남구", "강남").replace('서초구','서초').replace('관악구','관악').replace('여의도','여의')

data
```

```python
#측정일이 object로 찍히는데 이거를 날짜 단위로 핸들링할 때
print(data['측정일'])
data['측정일'] = pd.to_datetime(data['측정일'])
print(data['측정일'])
data['측정일'].dt.month

#날짜형이면 달, 날, 요일, 주 다 알 수 있다.
```

```python
data[data["담당"].isin(["박현경", "김현경"])]
```

```python
data.loc[data["담당"] == "김현경", ["몸무게"]]
```

```python
data.loc[data["담당"] == "김현경", ["이름", "측정일", "몸무게"]]
```

```python
## pivot_table :

pd.pivot_table(data, index=["지점", "담당"], values="몸무게", aggfunc='sum') #aggfunc 기본값: mean평균
```
