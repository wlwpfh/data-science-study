```python
import pandas as pd
import sqlite3
```

```python
connection=sqlite3.connect('pandas-transaction.db')

connection
```

### 메타 데이터 테이블 

```python
query="select * from 'sqlite_master'"
# mapping된 테이블 ( meta-table <- sw가 관리)

pd.read_sql(query, connection)
```

```python
#query= "select * from 'transaction'"

query="""
    select *
    from 'transaction'
    order by amount asc
    limit 2
"""

#head대신에 limit을 사용할 수 있음

transaction=pd.read_sql(query, connection)
transaction
```

```python

query="""
    select Name, amount
    from 'transaction'
    where name!='Kim'
    and amount<1000
    order by amount asc
"""

transaction=pd.read_sql(query, connection)
transaction
```

```python
query="select * from 'transaction' where product in ('cereal','kiwi')"
transaction=pd.read_sql(query, connection)
transaction
```

```python
query="""
    select *
    from 'transaction'
    where product like '%b%'
"""

transaction=pd.read_sql(query, connection)
transaction
```

```python
query="""
    select *
    from 'transaction'
"""
transaction=pd.read_sql(query, connection)
transaction
```

```python
query="""
    select *
    from 'user'
"""

user=pd.read_sql(query, connection)
user
```

```python
#조인 연결

df=transaction.merge(user, how='left')
df
# pandas를 이용
```

```python
query="""
    select *
    from 'transaction' as t
    left join 'user' as u
    on t.name = u.name
"""

pd.read_sql(query, connection)
```
