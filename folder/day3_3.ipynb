```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import folium
```

```python
from folium.plugins import MarkerCluster
```

```python
shop = pd.read_csv("소상공인시장진흥공단_상가(상권)정보_서울_202109.csv", encoding='utf-8')
```

```python
shop_coffee = shop[shop["상권업종중분류명"].str.contains("커피")]

shop_coffee
```

```python
shop_coffee_sb = shop_coffee[shop_coffee["상호명"].str.contains("스타벅스")]

shop_coffee_sb.shape
```

```python
map4 = folium.Map(location=[shop_coffee_sb['위도'].mean(), shop_coffee_sb['경도'].mean()], zoom_start=12)

marker_cluster = MarkerCluster().add_to(map4)
```

```python
for n in shop_coffee_sb.index:
    folium.Marker([shop_coffee_sb.loc[n, "위도"], shop_coffee_sb.loc[n, "경도"]],
                 icon=folium.Icon(color='green', icon='star')).add_to(marker_cluster)
    
map4
```
