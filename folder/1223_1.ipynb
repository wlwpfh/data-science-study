```python
# 영화 랭킹 정보 url을 GET 방식으로 요청
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
movie_ranking = requests.get(url)
movie_ranking
```

```python
soup=BeautifulSoup(movie_ranking.content, 'html.parser')

soup
```

```python
soup.find('a',title="티탄")
```

```python
soup.find_all('div',class_="tit3")
```

```python
len(soup.find_all('div',class_="tit3"))
```

```python
soup.find_all('div',class_="tit3")[0].text
```

```python
# for문을 이용해서 타이틀 뽑아내기 가능
for i in range(0,len(soup.find_all('div',class_="tit3"))):
    title=soup.find_all('div',class_="tit3")[i].text
    print(title)
```

```python
title_list = soup.select('div.tit3 > a')
title_list[0:5]    
```

```python
# 영화제목만 발췌하기 위해 for 반복문 활용 예:
count = len(title_list)       # List 에 들어간 데이터 갯수 확인 시, len 사용

for n in range(0,count):
    title_name = title_list[n].text
    print(f"{n+1}위 {title_name}")
```

```python
soup.find('li',class_='num1')

soup.text
```

```python
soup.find_all('li',class_="num1")[0].find_all("p")[2].text

# list구조이기 때문에 soup.find_all('li',class_="num1")이라고 하면 에러난다. 
```

```python
for i in range(1, 10):
    title_text=soup.find_all('li',class_=f"num{i}")[0].find_all("p")[2].text
    split_text=title_text.split("[도서]")
    print(split_text[1])
```

+) 추가로 다르게 하는 법 

```python
for n in range(1,11):
    t_num='num'+str(n)
    title=soup.find_all('li',class_=t_num)[0].find_all('p')[2].text.split("[도서] ")[1]
    auth=soup.find_all('li', class_=t_num)[0].find_all('p',class_='aupu')[0].text.split("저 |")[0]
    print(f"{n}위 도서명: {title}, 저자: {auth}")
    print("\n")
```
